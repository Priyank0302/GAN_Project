# 1. Matplotlib Introduction

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

It is highly customizable and widely used in data visualization.

Commonly used along with NumPy, pandas, and other libraries.

### Key Features:

Generates plots, bar charts, scatter plots, histograms, and pie charts.

Extensive style and customization options.

Supports output to multiple formats (PNG, PDF, SVG, etc.).

```python
# Creating Your First Plot
import matplotlib.pyplot as plt

# Basic plot example
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title("Basic Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

## Matplotlib Pyplot

Pyplot is a module in Matplotlib, providing a MATLAB-like interface.

Includes functions like plot(), show(), title(), xlabel(), ylabel(), etc.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, label="Line 1", color="blue", linestyle="--", linewidth=2)
plt.legend()
plt.show()
```

Matplotlib Plotting

Line Plots

Customization:

Line styles: '-', '--', '-.', ':'.

Colors: 'r', 'g', 'b', etc.

```python
import matplotlib.pyplot as plt

x = range(10)
y = [xi**2 for xi in x]

plt.plot(x, y, label="Quadratic")
plt.legend()
plt.show()
```

```python
# Adding Markers to Plots
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, marker='o', markersize=10, markerfacecolor='red', markeredgecolor='black')
plt.show()

# '.' (point), ',' (pixel), 'o' (circle), 'v' (triangle), '*' (star).

'''Additional Details:

markersize: Adjusts the size of the markers.

markerfacecolor: Sets the fill color of the marker.

markeredgecolor: Specifies the edge color of the marker.'''
```

```python
plt.plot(x, y, 'ro-') # red circle markers with line ''marker|line|color''
```

```python
plt.plot(x, y, color='blue', linewidth=2.5, linestyle='--')
plt.show()
```

Parameters:

color: Sets the line color (e.g., 'blue', '#FF5733').

linewidth: Adjusts the thickness of the line.

linestyle: Modifies the style of the line ('-', '--', ':', etc.).

```python
plt.plot(x, y)
plt.title("My Plot Title", fontsize=16)
plt.xlabel("X Axis", fontsize=14)
plt.ylabel("Y Axis", fontsize=14)
```

Additional Details:

fontsize: Sets the font size for the labels.

fontweight: Can be used to make the text bold (e.g., 'bold').

```python
plt.plot(x, y)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
```

Additional Details:

color: Specifies the color of the grid lines.

linestyle: Determines the style of the grid lines ('--', ':', etc.).

linewidth: Adjusts the thickness of the grid lines.

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y, 'r')
axs[0, 1].plot(x, [xi*2 for xi in x], 'g')
plt.show()
```

Additional Details:

subplots: Creates a grid of subplots within a figure.

figsize: Allows customization of the overall figure size (e.g., figsize=(10, 5)).

tight_layout: Ensures subplots fit neatly without overlapping.

```python
plt.scatter(x, y, c='blue', s=50, alpha=0.5)
plt.show()
```

Additional Details:

c: Sets the color of the points.

s: Adjusts the size of the points.

alpha: Sets the transparency level (0.0 to 1.0).

```python
plt.bar(x, y, color='cyan', edgecolor='black')
plt.show()
```

Additional Details:

color: Sets the fill color of the bars.

edgecolor: Specifies the color of the edges of the bars.

width: Adjusts the width of the bars.



```python
# Histograms
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins=4, color='purple', edgecolor='black')
plt.show()
```

Additional Details:

bins: Defines the number of bins (intervals) in the histogram.

color: Sets the fill color of the bars.

edgecolor: Specifies the color of the edges of the bars.

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.show()
```

Additional Details:

autopct: Adds percentage values to the slices. E.g., '%1.1f%%' formats to one decimal place.

edgecolor: Sets the edge color for each slice.

startangle: Rotates the start position of the pie chart (e.g., 90, 140).
